```python
 import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv("C:\\Users\\sandhya\\Downloads\\disney_plus_titles.csv")
```

```python
# Set a style
sns.set_style("whitegrid")
```

```python
genre_count = data['listed_in'].str.split(', ').explode().value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=genre_count.head(10).index, y=genre_count.head(10).values,
hue=genre_count.head(10).index, palette="viridis", dodge=False)
plt.legend([],[], frameon=False)
plt.title('Top 10 Most Frequent Genres on Disney Plus', fontsize=16)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
country_genre_count =data['country'].str.split(', ').explode().value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=country_genre_count.head(10).index, y=country_genre_count.
head(10).values, hue=country_genre_count.head(10).index, palette="magma",
dodge=False)
plt.legend([],[], frameon=False)
plt.title('Top 10 Countries Producing Content on Disney Plus', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Number of Titles', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
data['date_added'] = pd.to_datetime(data['date_added'])
rating_trend = data.groupby([data['date_added'].dt.year, 'rating']).size().unstack().fillna(0)
plt.figure(figsize=(13, 7))
sns.lineplot(data=rating_trend)
plt.title('Distribution of Content Ratings Over Time', fontsize=17)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Number of Titles', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(title='Rating', fontsize=13, title_fontsize=15)
plt.show()
```

```python
director_count = data['director'].value_counts().head(10)
plt.figure(figsize=(13, 7))
sns.barplot(x=director_count.index, y=director_count.values, hue=director_count.
index, palette="cubehelix", dodge=False)
plt.legend([],[], frameon=False)
plt.title('Top 10 Most Prolific Directors on Disney Plus', fontsize=17)
plt.xlabel('Director', fontsize=15)
plt.ylabel('Number of Titles', fontsize=15)
plt.xticks(rotation=47, fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

```python
data['cast_size'] = data['cast'].apply(lambda x: len(x.split(', ')) if pd.notnull(x) else 0)
cast_size_distribution = data.groupby('type')['cast_size'].mean()
plt.figure(figsize=(13, 7))
sns.barplot(x=cast_size_distribution.index, y=cast_size_distribution.values,
hue=cast_size_distribution.index, palette="muted", dodge=False)
plt.legend([],[], frameon=False)
plt.title('Average Cast Size of Movies vs. TV Shows on Disney Plus',
fontsize=16)
plt.xlabel('Type', fontsize=15)
plt.ylabel('Average Cast Size', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

```python
data['duration_minutes'] = data['duration'].apply(lambda x: int(x.split()[0])
if 'min' in x else None)
average_duration = data.groupby('type')['duration_minutes'].mean()
plt.figure(figsize=(13, 7))
sns.barplot(x=average_duration.index, y=average_duration.values,
hue=average_duration.index, palette="pastel", dodge=False)
plt.legend([],[], frameon=False)
plt.title('Average Duration of Movies vs. TV Shows on Disney Plus', fontsize=17)
plt.xlabel('Type', fontsize=14)
plt.ylabel('Average Duration (minutes)', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

```python
 plt.figure(figsize=(15, 11))
corr_matrix = data[['duration_minutes', 'cast_size']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix', fontsize=17)
plt.xticks(fontsize=13, rotation=46)
plt.yticks(fontsize=13)
plt.show()
```

```python
 plt.figure(figsize=(13, 7))
sns.histplot(data[data['type'] == 'Movie']['duration_minutes'].dropna(), bins=30, kde=True, color='blue')
plt.title('Distribution of Movie Durations', fontsize=17)
plt.xlabel('Duration (minutes)', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(13, 7))
sns.boxplot(x='type', y='duration_minutes', data=data)
plt.title('Boxplot of Duration by Content Type', fontsize=17)
plt.xlabel('Content Type', fontsize=15)
plt.ylabel('Duration (minutes)', fontsize=15)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```
